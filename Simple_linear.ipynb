# simple linear reg model 

```python
import statsmodels.api as sn
```

```python
X = sn.add_constant(df['room_num'])
```

```python
lm = sn.OLS(df['price'],X).fit()
```

1. lm = linear model
2. OLS = Oridnary least squares
3. dependent var, indep var

```python
lm.summary()
```

1. const b0, room_num b1
2. S.E. of b0 and b1
 * p < 1% or 5% for sig relationship bw x and y
 * confidence interval
3. r2 and adj r2 (close to 0.5)
